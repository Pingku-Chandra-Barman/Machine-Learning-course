```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=df.copy()
df2=df.copy()
df3=df.copy()
df4=df.copy()
```

# Scaling

```python
# Perform scaling on ['Marketing Spend', 'Administration', 'Transport'] seperately
```

# 1.Normalization


```python
df1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(df1[['Marketing Spend']])
df1.head()
```

```python
df1['Marketing Spend']=scaler.transform(df1[['Marketing Spend']])
df1.head()
```

```python
scaler.fit(df1[['Administration']])
df1['Administration']=scaler.transform(df1[['Administration']])
df1.head()
```

```python
scaler.fit(df1[['Transport']])
df1['Transport']=scaler.transform(df1[['Transport']])
df1.head()
```

# 2.Standardization


```python
df2.head()
```

```python
from sklearn.preprocessing import StandardScaler
stds = StandardScaler()
```

```python
df2['Marketing Spend']=stds.fit_transform(df2[['Marketing Spend']])
df2.head()
```

```python
df2['Administration']=stds.fit_transform(df2[['Administration']])
df2.head()
```

```python
df2['Transport']=stds.fit_transform(df2[['Transport']])
df2.head()
```
