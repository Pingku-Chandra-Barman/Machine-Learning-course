# One Hot Encoding

```python
df6.head()
```

```python
dummy=pd.get_dummies(df6['sex'],columns=df6.sex.unique)
dummy.head()
```

```python
df6.drop('sex',axis=1,inplace=True)
df6.head()
```

```python
df6=pd.concat([df6,dummy],axis=1)
df6.head()
```

```python
dummy1=pd.get_dummies(df6['smoker'],columns=df.smoker.unique())
dummy1.head()
```

```python
df6.drop('smoker',axis=1,inplace=True)
df6=pd.concat([df6,dummy1],axis=1)
df6.head()
```

```python
dummy2=pd.get_dummies(df6['region'],columns=df.region.unique(),drop_first=True,prefix='region')
dummy2.head()
```

```python
df6.drop('region',axis=1,inplace=True)
df6=pd.concat([df6,dummy2],axis=1)
df6.head()
```

# One hot encoding using loop

```python
df7.head()
```

```python
cols = ['sex','smoker','region']
for col in cols:
    one = pd.get_dummies(df8[col])
    df8 = pd.concat([df8, one], axis=1).drop(col, axis=1)
```

# Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df9.head()
```

```python
sex=df9.sex.unique()
sex
```

```python
ord1=OrdinalEncoder(categories=[sex])
```

```python
sex1=ord1.fit_transform(df9[['sex']])
#sex1
```

```python
sex1 = pd.DataFrame(sex1, columns=['sex'])
```

```python
df9 = pd.concat([df9.drop('sex', axis=1), sex1], axis=1)
df9.head()
```

```python
smoker=df9.smoker.unique()
smoker
```

```python
ord2=OrdinalEncoder(categories=[smoker])
```

```python
smoker1=ord2.fit_transform(df9[['smoker']])
smoker1 = pd.DataFrame(smoker1, columns=['smoker'])
df9 = pd.concat([df9.drop('smoker', axis=1), smoker1], axis=1)
df9.head()
```

```python
region=df9.region.unique()
ord3=OrdinalEncoder(categories=[region])
region1=ord3.fit_transform(df9[['region']])
region1 = pd.DataFrame(region1, columns=['region'])
df9 = pd.concat([df9.drop('region', axis=1), region1], axis=1)
df9.head()
```
